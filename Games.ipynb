## Предобработка данных

**Приведем названия столбцов к нижнему регистру**

```python
df.columns = df.columns.str.lower()
```

**Для заполнения пропусков в столбцах `year_of_release`, `rating` воспользуемся функцией. Будем заменять существующими значениями идентичных игр с других платформ, если таковые имеются. Предварительно пропущенные значения `rating` заменим текстовыми значениями `none` - необходимо для корректного сравнения строк формата `str`.**

```python
df['rating'] = df['rating'].fillna('none')
```

```python
def table_fillna(row):
    if pd.isna(row['year_of_release']):
        row['year_of_release'] = df[df['name'] == row['name']]['year_of_release'].min()
#     if row['rating'] == 'none':
#         row['rating'] = df[df['name'] == row['name']]['rating'].min()
    return row
```

```python
df = df.apply(table_fillna, axis=1)
```

```python
# оставшееся кол-во игр с незаполненным годом выхода.
df[pd.isna(df['year_of_release'])]['platform'].count()
```

```python
# оставшееся кол-во игр с незаполненным названием.
df[pd.isna(df['name'])]['platform'].count()
```

**Удалим строки с оставшимися пропусками в столбцах: год выхода `146`шт., название игры, всего `2шт`.**

```python
df.dropna(subset=['name', 'year_of_release'], inplace=True)
```

**Приведем столбец `year_of_release` к целочисленному типу данных.**

```python
df['year_of_release'] = df['year_of_release'].astype('int')
```

**Заменим в столбце `'user_score'` значения `'tbd'` пропусками `np.nan` и приведем столбец к числовому формату.**

```python
df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan
df['user_score'] = pd.to_numeric(df['user_score'])
```

**В столбце возрастного рейтинга объединим категории `'K-A'` и `E`, оставим только `E`.**

```python
df.loc[df['rating'] == 'K-A', 'rating'] = 'E'
```

```python
df['rating'].unique()
```

**Создадим столбец `'total_sales'` с суммой кол-ва проданных копий по всем регионам.**

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales']
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**Для проверки гипотезы `Средние пользовательские рейтинги платформ Xbox One и PC одинаковые` воспользуемся методом t-теста для проверки гипотезы о равенстве среднего двух совокупностей. Данный метод был выбран, так как нам необходимо сравнить между собой независимые между собой выборки.**


**Нулевая гипотеза: пользовательские рейтинги платформ Xbox One и PC одинаковые.** 

**Альтернативная: пользовательские рейтинги платформ Xbox One и PC не равны.**


**Подготовим массивы для расчетов, удалим пропущенные значения, посмотрим на размерность. А также выделим таблицы, чтобы построить гистограммы для наглядности.**

```python
xbox_one_table = df[df['platform'] == 'XOne']['user_score']
xbox_one = np.array(df[df['platform'] == 'XOne']['user_score'].dropna())
xbox_one.shape
```

```python
pc_table = df[df['platform'] == 'PC']['user_score']
pc = np.array(df[df['platform'] == 'PC']['user_score'].dropna())
pc.shape
```

**Сравним значения дисперсий.**

```python
np.var(xbox_one)
```

```python
np.var(pc)
```

**Размерность вполне высокая для проведения теста. За параметр alpha примем общепринятое значение в `0.05`, параметр разности дисперсий опустим.**

```python
alpha = 0.05
```

```python
platform_score_p = st.ttest_ind(xbox_one, pc)
```

```python
platform_score_p.pvalue
```

```python
if platform_score_p.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Таким образом отвергаем 0-вую гипотезу. Это значит, что показатели средней выборки по отзывам Xbox различаются со средней выборкой по пользовательским отзывам платформы PC. Принимаем альтернативную гипотезу. Для наглядности взглянем на гистограммы:**

```python
ax = xbox_one_table.plot(
    kind='hist',
    y='user_score',
    histtype='step',
    range=(0, 11),
    bins=20,
    linewidth=5,
    alpha=0.7,
    label='Пользовательские оценки Xbox One',
    legend=True)
    
pc_table.plot(
    kind='hist',
    y='user_score',
    histtype='step',
    range=(0, 11),
    bins=20,
    linewidth=5,
    alpha=0.7,
    label='Пользовательские оценки PC',
    ax=ax,
    legend=True)
    
plt.title('Пользовательские оценки платформ', fontsize=16)
plt.grid()
plt.xticks(np.arange(11, step=1))
plt.xlabel('Оценка', fontsize=12)
plt.ylim(0, 175)
plt.show()
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

**Для проверки гипотезы `"Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные"` также воспользуемся методом t-теста для проверки гипотезы о равенстве среднего двух совокупностей.**

**Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые.** 

**Альтернативная: средние пользовательские рейтинги жанров Action и Sports разные.**

```python
# создадим таблицы для гистограмм, массивы для расчетов
action_table = df[df['genre'] == 'Action']
action = np.array(df[df['genre'] == 'Action']['user_score'].dropna())
action.shape
```

```python
sports_table = df[df['genre'] == 'Sports']
sports = np.array(df[df['genre'] == 'Sports']['user_score'].dropna())
sports.shape
```

```python
np.var(action)
```

```python
np.var(sports)
```

**Значений для расчета достаточно, показатели дисперсий примерно одинаковы, за alpha также примем `0.05`, как общепринятое значение.**

```python
genre_score_p = st.ttest_ind(action, sports)
```

```python
genre_score_p.pvalue
```

```python
if genre_score_p.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Принимаем нулевую гипотезу, пользовательские отзывы по играм жанра `Action` схожи с отзывами жанра `Sport`, взглянем на гистограммы.**
